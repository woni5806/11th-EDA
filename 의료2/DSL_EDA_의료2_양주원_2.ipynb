```python
#KOSIS 데이터
fd_d = pd.read_csv('/content/drive/MyDrive/Data/미세먼지_Kosis.csv' , encoding='cp949') #미세먼지 데이터
ozon = pd.read_csv('/content/drive/MyDrive/Data/오존(10월)포함.csv' , encoding='cp949')
```

```python
fd_d.head
```

```python
# '단위' 열과 마지막 열 'Unnamed' 열을 제거
fd_d.drop(columns='단위', inplace=True)
fd_d.head(3)
```

```python
# melt 함수를 사용하여 데이터 재구성
# '구분' 열을 id_vars로 지정하여 지역 정보를 유지
# 나머지 열들은 연도와 월 정보로 변환됨
df_melted = pd.melt(fd_d, id_vars=['구분'], var_name='날짜', value_name='값', value_vars=fd_d.columns[2:])

# '날짜' 열에서 연도와 월을 분리
df_melted['연도'] = df_melted['날짜'].apply(lambda x: x.split('.')[0])
df_melted['월'] = df_melted['날짜'].apply(lambda x: x.split('.')[1].split()[0])
```

```python
df_melted
```

```python
# DataFrame을 CSV 파일로 저장하기
df_melted.to_csv('/content/drive/MyDrive/Data/미먼 데이터.csv', index=False, encoding='cp949')
```

```python
ozon.head()
```

```python
from datetime import datetime

# 'Date' 열을 datetime으로 변환하고 연도와 월 추출
ozon['Year'] = ozon['일시'].apply(lambda x: datetime.strptime(x, '%y.%b').year)
ozon['Month'] = ozon['일시'].apply(lambda x: datetime.strptime(x, '%y.%b').month)

ozon.head()
```

```python
#df_melted의 'Year', 'Month' 열을 정수형으로 변환
df_melted['Year'] = df_melted['Year'].astype(int)
df_melted['Month'] = df_melted['Month'].astype(int)
```

```python
# 병합을 위해 df_melted 데이터프레임의 '연도', '월' 열 이름을 'Year', 'Month'로 변경
df_melted.rename(columns={'연도': 'Year', '월': 'Month', '구분' : '주소(시도)'}, inplace=True)

# ozon 데이터프레임과 df_melted 데이터프레임을 'Year'와 'Month' 열을 기준으로 병합
merged_df = pd.merge(ozon, df_melted, on=['Year', 'Month', '주소(시도)'], how='left')

merged_df[merged_df['주소(시도)'] == '제주']
```

```python
# 병합된 데이터프레임(예를 들어 merged_df라고 가정)에서 열을 제거
merged_df = merged_df.drop(['Year', 'Month', '날짜', '미세먼지'], axis=1)

# 결과 확인
merged_df.head()
```

```python
# DataFrame을 CSV 파일로 저장하기
merged_df.to_csv('/content/drive/MyDrive/Data/최종 미먼 data.csv', index=False, encoding='cp949')
```
